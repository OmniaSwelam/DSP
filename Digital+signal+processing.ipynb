```python
import numpy as np
import matplotlib.pyplot as plt
#import matplotlib
#matplotlib.use("TKAgg")
from matplotlib import style
from tkinter import *
#import tkinter as Tk
from tkinter import filedialog
import pandas
import scipy
from scipy import signal
import pyaudio
p = pyaudio.PyAudio()
```

```python
#window (filter) variables
global gA, gB, gC, gD, gE
#sound signals variables
global g_f1, g_f2, g_f3, g_lamda
#Sampling signals variables
global g_fs, g_f


root= Tk()
root.title('DSP')
root.geometry('600x300')


label= Label(root, text= 'Window', bg='black', fg= 'white').grid(row= 1, columnspan= 3)

#file Browser from desktop
def Insert():
    global signal
    root.fileName= filedialog.askopenfilename(filetypes={("csv files", "*.csv") })
    name= root.fileName
    print(name)
    signal = pandas.read_csv(name)
    plt.plot(signal['0'])
    plt.show()
    return signal
insert= Button(root, text='Insert', bg='purple',fg='white', command=Insert)
insert.grid(row=0, column=1)

#filtration - Convlution
a= Entry(root)
a.grid(row=3, column=0)
a.delete(0,END)
a.insert(0,' ')


b= Entry(root)
b.grid(row=3, column=1)
b.delete(0,END) #delete
b.insert(0,' ') #default value


c= Entry(root) 
c.grid(row=3, column=2)
c.delete(0,END)
c.insert(0,' ')


d= Entry(root)
d.grid(row=3, column=3)
d.delete(0,END)
d.insert(0,' ')


e= Entry(root)
e.grid(row=3, column=4)
e.delete(0,END)
e.insert(0,' ')

#Sound system (sins and exponential)
soundlabel= Label(root, text='Sound Signals - Input Freq.', fg='purple',bg='white').grid(row=6, column= 0)
f1= Entry(root)
f1.grid(row=7, column=0)

f2= Entry(root)
f2.grid(row=8, column=0)

f3= Entry(root)
f3.grid(row=9, column=0)

lamda=Entry(root)
lamda.grid(row=10,column=0)

label_f1= Label(root, text= "Sin 1 Freq.")
label_f1.grid(row=7, column=1, sticky= W)

label_f2= Label(root, text= "Sin 2 Freq.")
label_f2.grid(row=8, column=1, sticky= W)

label_f1= Label(root, text= "Sin 3 Freq.")
label_f1.grid(row=9, column=1, sticky= W)

label_lamda= Label(root, text= "exp")
label_lamda.grid(row=10, column=1, sticky= W)
#Sampling gui
fs= Entry(root)
fs.grid(row=12,column=1)
f= Entry(root)
f.grid(row=13,column=1)

label_fs= Label(root, text= 'Fs')
label_fs.grid(row=12, column=0)
label_fs= Label(root, text= 'F')
label_fs.grid(row=13, column=0)

def Filter():
    gA= eval(a.get())
    gB= eval(b.get())
    gC= eval(c.get())
    gD= eval(d.get())
    gE= eval(e.get())
    #print gA, gB, gC, gD, gE
    window= np.array([gA, gB, gC, gD, gE])
    print(window.shape)
    #print signal.shape
    #sig = np.asarray(signal)
    #print signal['0']
    print(signal['0'].shape)
    new_signal= scipy.signal.convolve(signal['0'],window, mode='same')
    #print new_signal
    print(new_signal.shape)
    plt.plot(new_signal)
    plt.show()
    return new_signal

    
filter= Button(root, text='Filter', bg='purple',fg='white', command= Filter)
filter.grid(row=1, column=2)

    
def listen_sound():
    volume =0.5
    fs=44100
    duration = 10.0
    g_f1= eval(f1.get())
    g_f2= eval(f2.get())
    g_f3= eval(f3.get())
    g_lamda= eval(lamda.get())
    
    samples= (np.sin(2*np.pi*np.arange(fs*duration)*g_f1/fs)).astype(np.float32)
    +(np.sin(2*np.pi*np.arange(fs*duration)*g_f2/fs)).astype(np.float32) 
    +(np.sin(2*np.pi*np.arange(fs*duration)*g_f3/fs)).astype(np.float32)
    +(np.exp(np.arange(fs*duration)*g_lamda/fs)).astype(np.float32)
    
    stream = p.open(format=pyaudio.paFloat32,
                 channels=1,
                rate=fs,
                output=True)
    stream.write(volume*samples)
    stream.stop_stream()
    stream.close()
    p.terminate()
    
sound= Button(root, text= 'Sound', bg='purple', fg='white',command= listen_sound)
sound= sound.grid(row=11,column=0)
def sample_sin():
    g_fs=eval(fs.get())
    g_f= eval(f.get())
    x = np.arange(g_fs)
    #print(x)# the points on the x axis for plotting
    # compute the value (amplitude) of the sin wave at the for each sample
    y = [ np.sin(2*np.pi*g_f * (i/g_fs)) for i in x]

    #this instruction can only be used with IPython Notbook. 
    % matplotlib inline
    # showing the exact location of the smaples
    plt.subplot(1,2,1)
    plt.plot(x,y)
    plt.title('before sampling')
    #plt.xlim(0,10)
    plt.subplot(1,2,2)
    plt.stem(x,y, 'b', )
    plt.title('after sampling')
    plt.show()

    
def sample_exp():
    g_fs=eval(fs.get())
    g_f= eval(f.get())
    x = np.arange(g_fs) # the points on the x axis for plotting
    # compute the value (amplitude) of the sin wave at the for each sample
    y = [ np.exp(-g_f* (i/g_fs) ) for i in x]

    #this instruction can only be used with IPython Notbook. 
    % matplotlib inline
    # showing the exact location of the smaples
    plt.subplot(1,2,1)
    plt.plot(x,y)
    plt.title('before sampling')
    plt.subplot(1,2,2)
    plt.stem(x,y, 'b', )
    plt.title('after sampling')
    plt.show()
sample1= Button(root, text= 'Sample sin', command= sample_sin, bg= 'purple', fg='white')
sample1.grid(row=14, columnspan=2)
sample2= Button(root, text= 'Sample exp', command= sample_exp, bg='purple', fg='white')
sample2.grid(row=15, columnspan=2)

root.mainloop()
```
